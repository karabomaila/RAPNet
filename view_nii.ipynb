### Display only the first 6 slices of the output mask image

```python
plt.subplots(2, 3, figsize=(8, 8))
for i in range(6):
    im = prediction_mask.get_fdata()
    arr = im[:, :, i]
    
    plt.subplot(2, 3, i + 1)
    plt.title(f"Slice number {i}")
    plt.imshow(arr, cmap='gray')
plt.tight_layout()
plt.show()
```
